## 4: Sampling from the Business Data


```python
def sample(series, n):

    np.random.seed(40)
    chosen_list = np.random.choice(series, size = n, replace = False)
    return list(chosen_list)
    
```

Collect a stratified random sample of business names, where each stratum consists of a postal code. Collect one business name per stratum. 

```python
bus.head()
```

```python
strats = bus['name'].groupby(bus['postal_code_5'])
```

```python
stratum = bus.groupby('postal_code_5').apply(lambda x: x.sample(1)).reset_index(drop=True)
stratum
```

```python
bus_strat_sample = stratum['name']
bus_strat_sample.head()
```

Collect a cluster sample of business IDs, where each cluster is a postal code, with 5 clusters in the sample. Assign `bus_cluster_sample` to a series of business IDs selected by this sampling procedure.



```python
bus.head()
```

```python
def sample(series, n):
    np.random.seed(40)
    chosen_list = np.random.choice(series, size = n, replace = False)
    return list(chosen_list)
```

```python
d = []
for i in bus['business_id'].groupby(bus['postal_code_5']).groups:
    d.append(i)
c = sample(d,5)
c

```

```python
bus_cluster_sample = bus[bus['postal_code_5'].isin(c)]
bus_cluster_sample.head()
```

*While cluster sampling seems like it will be less costly in terms of computation time requirement, there is also a chance to miss out on understanding the entire picture when using this sampling technique as only some of the groups are selected. And while elements inside a group might have diverse behaviors, several different groups might exist completely outside the range of these selected groups. In stratified sampling, all groups are considered which ensures that it might be a better representation sample of the entire subset*

Collect a multi-stage sample. First, take a SRS of 5 postal codes. Then, collect an SRS of one business name per selected postal code.

```python
np.random.seed(40) 
```

```python
bus_multi_sample = bus[bus['postal_code_5'].isin(np.random.choice(bus['postal_code_5'].unique(),size=5,replace=False))].groupby('postal_code_5')['name'].agg(lambda x: sample(x, 1)[0]) # SOLUTION
bus_multi_sample.head()
```

## 8: Restaurant Ratings Over Time

Let's consider various scenarios involving restaurants with multiple ratings over time.

Let's see which restaurant has had the most extreme improvement in its rating. 

```python
ins_named.head()
```

```python
t1 = ins_named['business_id'].value_counts() - 3
t1.head()
```

```python
t1 = pd.DataFrame(t1.rename('Number of Inspections'))
t2 = t1[t1['Number of Inspections'] >= 0]
t2['business_id'] = t2.index
t3 = t2['business_id'].unique()

```

```python
def swing(idd):
    d = ins[ins['business_id'] == idd]
    k = d['score'].max()
    l = d['score'].min()
    return(k-l,idd)

allswing = []
bid = []
for i in t3:
    g,h = swing(i)
    allswing.append(g)
    bid.append(h)
```

```python
d={'Swing': allswing, 'business_id': bid}

j = pd.DataFrame(data = d)
print(j['Swing'].max())

x = j[j['Swing'] == j['Swing'].max()]['business_id']
print(x)
```


What's the relationship between the first and second scores for the businesses with 2 inspections in a year? Do they typically improve? For simplicity, let's focus on only 2016 for this problem.



```python
# Create the dataframe here
scores_pairs_by_business = scores_pairs_by_business = (ins2016.sort_values('date')
                            .loc[:, ['business_id', 'score']]
                            .groupby('business_id')
                            .filter(lambda group: len(group)==2)
                            .groupby('business_id')
                            .agg(list)
                            .rename(columns={'score':'score_pair'}))
ins2016 = ins[ins['year'] == 2016]
```

```python
import matplotlib.pyplot as plt

first_score, second_score = zip(*scores_pairs_by_business['score_pair'])
plt.scatter(first_score,second_score,s=20,facecolors='none',edgecolors='b')
plt.plot([55,100],[55,100],'r-')
plt.xlabel('First Score')
plt.ylabel('Second Score')
plt.axis([55,100,55,100])
plt.title("First Inspection Score vs. Second Inspection Score");
```


Another way to compare the scores from the two inspections is to examine the difference in scores. 

```python
diffs = np.array(second_score) - np.array(first_score)
plt.hist(diffs, bins=30)
plt.title("Distribution of Score Differences")
plt.xlabel("Score Difference (Second Score - First Score)")
plt.ylabel("Count");
```

If the restaurants tend to improve from the first to the second inspection, 
we would expect to see the points in the scatter plot fall above the line of slope 1. 
We would also expect to see the histogram of the difference in scores to be shifted toward
positive values. Interestingly, we don't see this. The second inspection often is worse than first. 
The histogram of differences shows a unimodal distribution centered at 0, hinting that the
average restaurant does not see a change in score between their first and second inspection.
This distribution has long tails with some scores being as low as -20 and others as high as 20-30.
